```python
import sqlite3
from sqlite3 import Error

def create_connection(path):
    connection = None
    try:
        connection = sqlite3.connect(path)
        print("Connection to SQLite DB successful")
    except Error as e:
        print(f"The error '{e}' occurred")

    return connection

def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query executed successfully")
    except Error as e:
        print(f"The error '{e}' occurred")

connection = create_connection("review.sqlite")
```

```python
create_articles_table = """
CREATE TABLE IF NOT EXISTS articles (
  pmid INTEGER,
  title TEXT,
  authors TEXT,
  first_author TEXT,
  journal TEXT,
  publication_year YEAR,
  create_date DATE,
  nihms_id TEXT,
  pmcid TEXT,
  doi TEXT,
  of_interest BOOL
);
"""
execute_query(connection, create_articles_table)

create_abstracts_table = """
CREATE TABLE IF NOT EXISTS abstracts (
  abstract TEXT,
  pmid INTEGER,
  FOREIGN KEY (pmid) REFERENCES articles (pmid)
);
"""
execute_query(connection, create_abstracts_table)
```

```python
create_column = """
ALTER TABLE articles
ADD citation TEXT;
"""
execute_query(connection, create_column)

create_column = """
ALTER TABLE articles
ADD pmcid TEXT;
"""
execute_query(connection, create_column)
```

```python
for row in range(len(df)):
    insert_article = f"""
    INSERT INTO
        articles (pmid, title, authors, citation, first_author, journal, publication_year, create_date, pmcid, nihms_id, doi, of_interest)
    VALUES
        ({df.iloc[row]['PMID']}, quote("{df.iloc[row]['Title'].replace('"',"'")}"), "{df.iloc[row]['Authors']}", "{df.iloc[row]['Citation']}", "{df.iloc[row]['First Author']}", "{df.iloc[row]['Journal/Book']}", {df.iloc[row]['Publication Year']}, {df.iloc[row]['Create Date']}, "{df.iloc[row]['PMCID']}", "{df.iloc[row]['NIHMS ID']}", "{df.iloc[row]['DOI']}", NULL);
    """
    execute_query(connection, insert_article)

    insert_abstract = f"""
    INSERT INTO
        abstracts (pmid, abstract)
    VALUES
        ({df.iloc[row]['PMID']}, quote("{abstracts[row].replace('"',"'")}"));
    """
    execute_query(connection, insert_abstract)
```
